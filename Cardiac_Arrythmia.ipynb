```
##As we can see there are outliers present in most of the columns, we will scale those outliers to the Q1-1.5*IQR and Q3+1.5*IQR values.
```

```
df3=df.drop('class', axis=1)
```

```
for name in list(df3.columns):
  upper_quartile = np.percentile(df3[name],75)
  lower_quartile = np.percentile(df3[name],25)
  IQR = (upper_quartile - lower_quartile)
  quartileSet = (lower_quartile - 1.5*IQR, upper_quartile + 1.5*IQR)
  for i in range(0,400):
      if df3[name][i]<=quartileSet[0]:
        df3[name][i]=quartileSet[0]
      elif df3[name][i]>=quartileSet[1]:
        df3[name][i]=quartileSet[1]
      else:
        df3[name][i]
# Now, all the outliers have been removed
```

```
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```
#x=pd.concat([df,df2], axis=1)
x=df3
y=df['class']
```

```
X_train,X_test, y_train, y_test=train_test_split(x,y,random_state=0)
```

```
sc=MinMaxScaler()
```

```
xc_train=sc.fit_transform(X_train)
xc_test=sc.transform(X_test)
```

#PCA 

```
pca=PCA(n_components=0.95)
```

```
x_reduced=pca.fit_transform(xc_train)
xt_reduced=pca.transform(xc_test)
```

```
np.sum(pca.explained_variance_ratio_)
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
```

#KNN Classifier

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
knn_clf = KNeighborsClassifier()

param_grid = {'n_neighbors' : [1,2,3,4,5,7,10]}

grid_search = GridSearchCV (knn_clf, param_grid, cv=5, scoring= 'precision_macro' , return_train_score=True)

grid_search.fit(x_reduced, y_train)
```

```
grid_search.best_params_
```

```
knn_clf = KNeighborsClassifier(n_neighbors=1)
knn_clf.fit(x_reduced, y_train)
```

```
from sklearn.metrics import precision_score
from sklearn import metrics
y_pred_train = knn_clf.predict(x_reduced)
y_pred_test = knn_clf.predict(xt_reduced)

knn_train_precision_score = precision_score(y_train, y_pred_train, average='macro')
knn_test_precision_score = precision_score(y_test, y_pred_test, average='macro')

print('Train Precision score: {}'
      .format(knn_train_precision_score))
print('Test Precision score: {}'
      .format(knn_test_precision_score))

metrics.confusion_matrix(y_test, y_pred_test)
```
